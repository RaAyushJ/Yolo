```python
# ==============================================
# 🚀 STEP 1: Mount Google Drive
# ==============================================
from google.colab import drive
drive.mount('/content/drive')  # Mount Google Drive

# Define folder where ESP32-CAM images are stored
esp32_folder = "/content/drive/MyDrive/ESP32-CAM"
output_folder = "/content/drive/MyDrive/ESP32-CAM-Processed"

# Create output folder if it doesn't exist
import os
os.makedirs(output_folder, exist_ok=True)

# ==============================================
# 🚀 STEP 2: Install YOLOv8 (Ultralytics)
# ==============================================
!pip install ultralytics
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow

# Load YOLOv8 Model (using the smallest, fastest version)
model = YOLO("yolov8n.pt")  # 'n' stands for nano (small and fast)

# ==============================================
# 🚀 STEP 3: List Images in ESP32-CAM Folder
# ==============================================
image_files = [os.path.join(esp32_folder, f) for f in os.listdir(esp32_folder) if f.endswith(('.jpg', '.png'))]

if not image_files:
    print("⚠️ No images found in ESP32-CAM folder. Please check your Google Drive path!")
else:
    print(f"✅ Found {len(image_files)} images.")

# ==============================================
# 🚀 STEP 4: Run YOLOv8 on Each Image
# ==============================================
for img_path in image_files:
    print(f"🔍 Processing: {img_path}")
    results = model(img_path)  # Perform object detection

    # Display results in Colab
    for result in results:
        output_img = result.plot()  # Get image with detections
        cv2_imshow(output_img)  # Show image in Colab

        # Save the processed image to Google Drive
        save_path = os.path.join(output_folder, os.path.basename(img_path))
        cv2.imwrite(save_path, output_img)
        print(f"✅ Saved: {save_path}")

print("🎯 Processing completed. All detected images are saved in:", output_folder)
```
